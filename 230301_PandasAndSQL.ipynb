##JOIN queries

```python
pd.read_sql("SELECT * FROM Movies", con)
```

```python
pd.read_sql("SELECT * FROM Votes", con)
```

```python
pd.read_sql("SELECT * \
            FROM Movies \
            JOIN Votes \
            ON Movies.id=Votes.id", con)
```

```python
pd.read_sql("SELECT Movies.id, Movies.title, Votes.vote_average \
            FROM Movies \
            JOIN Votes \
            ON Movies.id = Votes.id", con, index_col = "id")
```

```python
pd.read_sql("SELECT Movies.id, Movies.title, Votes.vote_average \
            FROM Movies \
            JOIN Votes \
            ON Movies.id = Votes.id \
            WHERE Votes.vote_average > 8", con, index_col = "id")
```

```python
pd.read_sql("SELECT Movies.id, Movies.title, Movies.budget, Votes.vote_average \
            FROM Movies \
            JOIN Votes \
            ON Movies.id = Votes.id \
            WHERE Votes.vote_average > 7.5 \
            ORDER BY Movies.budget DESC", con, index_col = "id")
```

##What are the most successful companies in terms of total revenue?

```python
con = sq3.connect('/content/drive/MyDrive/230222_Master 12 Advanced Projects/movies.db')
```

```python
pd.read_sql("SELECT * FROM Movies", con)
```

```python
pd.read_sql("SELECT * FROM Prod_comp", con)
```

```python
df = pd.read_sql("SELECT Prod_comp.prod_comp_id, Prod_comp.prod_comp_name, Movies.revenue, Movies.title \
                  FROM Prod_comp \
                  LEFT JOIN Movies \
                  ON Prod_comp.id = Movies.id", con)
df
```

```python
df.groupby('prod_comp_name').revenue.sum().sort_values(ascending = False)
```

##What companies created the Titanic?

```python
pd.read_sql("SELECT Prod_comp.prod_comp_name \
            FROM Prod_comp \
            LEFT JOIN Movies \
            ON Prod_comp.id = Movies.id \
            WHERE Movies.title = 'Titanic'", con)
```

##What are the most successful genres in terms of total revenue?

```python
df2 = pd.read_sql("SELECT Genres.id, Genres.genre_name, Movies.revenue, Movies.title \
                  FROM Genres \
                  LEFT JOIN Movies \
                  ON Genres.id = Movies.id", con)
df2
```

```python
df2.groupby('genre_name').revenue.sum().sort_values(ascending = False)
```

##What is/are the genre/s of the movie Avatar?

```python
pd.read_sql("SELECT Genres.genre_name \
            FROM Genres \
            LEFT JOIN Movies \
            ON Genres.id = Movies.id \
            WHERE Movies.title = 'Avatar'", con)
```
