##Importing and merging many files

```python
df_1880 = pd.read_csv('/content/drive/MyDrive/230222_Master 12 Advanced Projects/Project_06_Materials/names/yob1880.txt', header = None, names = ['Name', 'Gender', 'Count'])
df_1881 = pd.read_csv('/content/drive/MyDrive/230222_Master 12 Advanced Projects/Project_06_Materials/names/yob1881.txt', header = None, names = ['Name', 'Gender', 'Count'])
```

```python
pd.concat(objs = [df_1880, df_1881], axis = 0)
```

###Multiindex with "keys" parameter

```python
pd.concat(objs = [df_1880, df_1881], axis = 0, keys = [1880, 1881], names = ['Year'])
```

###Dropping the inner index level with droplevel()

```python
pd.concat(objs = [df_1880, df_1881], axis = 0, keys = [1880, 1881], names = ['Year']).droplevel(-1).reset_index()
```

###String replacement

```python
years = list(range(1880, 2019))
years[:5]
```

```python
url = '/content/drive/MyDrive/230222_Master 12 Advanced Projects/Project_06_Materials/names/'
```

```python
dataframes = []
for year in years:
  data = pd.read_csv(url+'yob{}.txt'.format(year), header = None, names = ['Name', 'Gender', 'Count'])
  dataframes.append(data)
```

```python
dataframes[-2:]
```

```python
len(dataframes)
```

```python
df = pd.concat(dataframes, axis = 0, keys = years, names = ['Year']).droplevel(-1).reset_index()
```

```python
df.info()
```

###Glob module - finding all the pathnames matching a specified pattern

```python
from glob import glob
```

```python
path = '/content/drive/MyDrive/230222_Master 12 Advanced Projects/Project_06_Materials/namesbystate/'
```

```python
glob(path+'A?.TXT')
```

```python
glob(path+'*.TXT')[-5:]
```

```python
filenames = glob(path+'*.TXT')
```

```python
len(filenames)
```

```python
dataframes = []
for name in filenames:
  df = pd.read_csv(name, header = None, names = ['State', 'Gender', 'Year', 'Name', 'Count'])
  dataframes.append(df)
```

```python
dataframes[-2:]
```

```python
len(dataframes)
```

```python
df = pd.concat(dataframes, ignore_index = True)
df
```

```python
df.info()
```

```python
df.to_csv('/content/drive/MyDrive/230222_Master 12 Advanced Projects/us_baby_names_state.csv', index = False)
```
