##DJI

```python
data = pd.read_csv('/content/drive/MyDrive/230222_Master 12 Advanced Projects/dji.csv', parse_dates = ['Date'], index_col = 'Date')
```

```python
data.info()
```

```python
df = data.loc['2010-01-01': '2022-12-31', 'Close'].to_frame()
```

```python
df
```

```python
df.plot(figsize = (12, 6), fontsize = 15)
plt.legend(fontsize = 15)
plt.show()
```

###pct_change()

```python
df['Return'] = df.pct_change()
```

```python
df
```

```python
df.dropna(inplace = True)
```

```python
df.plot(figsize = (12, 6), secondary_y = 'Return', mark_right = True, fontsize = 15)
plt.show()
```

```python
df.rename(columns = {'Close': 'DJI_Close', 'Return': 'DJI_Return'}, inplace = True)
```

```python
df
```

##Backtesting a momentum strategy

```python
df['Position'] = np.sign(df['DJI_Return'])
```

```python
df
```

```python
df['Strategy_ret'] = df['Position'].shift() * df['DJI_Return']
```

```python
df
```

```python
df.Strategy_ret.add(1, fill_value = 0).cumprod()
```

```python
df['Strategy'] = df.Strategy_ret.add(1, fill_value = 0).cumprod() * df.iloc[0, 0]
```

```python
df
```

###Visualization

```python
df[['DJI_Close', 'Strategy']].plot(figsize = (12, 6), fontsize = 15)
plt.title('Simple momentum strategy', fontsize = 20)
plt.legend(fontsize = 15)
plt.show()
```

##Contrarian strategy

```python
df = data.loc["2010":, 'Close'].to_frame()
df['DJI_Return'] = df.Close.pct_change()
df.columns = ['DJI_Close', 'DJI_Return']
df.dropna(inplace = True)
df
```

```python
df['Position'] = -np.sign(df['DJI_Return'])
```

```python
df['Strategy_ret'] = df['Position'].shift() * df['DJI_Return']
df['Strategy'] = df.Strategy_ret.add(1, fill_value = 0).cumprod() * df.iloc[0, 0]
```

```python
df[['DJI_Close', 'Strategy']].plot(figsize = (12, 6), fontsize = 15)
plt.title('Simple contrarian strategy', fontsize = 20)
plt.legend(fontsize = 15)
plt.show()
```

##More complex strategies

```python
df = data.loc["2010":, 'Close'].to_frame()
df['DJI_Return'] = df.Close.pct_change()
df.columns = ['DJI_Close', 'DJI_Return']
df.dropna(inplace = True)
```

```python
df['Position'] = np.where(df['DJI_Return'] > 0.005, -1, 1)
```

```python
df['Strategy_ret'] = df['Position'].shift() * df['DJI_Return']
df['Strategy'] = df.Strategy_ret.add(1, fill_value = 0).cumprod() * df.iloc[0, 0]
```

```python
df[['DJI_Close', 'Strategy']].plot(figsize = (12, 6), fontsize = 15)
plt.title('Simple contrarian strategy', fontsize = 20)
plt.legend(fontsize = 15)
plt.show()
```

##Simple moving averages (SMA)

```python
df = data.loc["2010":, 'Close'].to_frame()
df['DJI_Return'] = df.Close.pct_change()
df.columns = ['DJI_Close', 'DJI_Return']
df.dropna(inplace = True)
```

```python
df.DJI_Close.rolling(window = 50).mean()
```

```python
df['SMA50'] = df.DJI_Close.rolling(window = 50).mean()
```

```python
df[['DJI_Close', 'SMA50']].plot(figsize = (12, 6), fontsize = 15)
plt.title('SMA50', fontsize = 20)
plt.legend(fontsize = 15)
plt.show()
```

```python
df['SMA200'] = df.DJI_Close.rolling(window = 200).mean()
```

```python
df[['SMA50', 'SMA200']].plot(figsize = (12, 6), fontsize = 15)
plt.title('SMA200', fontsize = 20)
plt.legend(fontsize = 15)
plt.show()
```

```python
df.dropna(inplace = True)
```

```python
df
```

```python
df['Position'] = np.sign(df.SMA50.sub(df.SMA200))
```

```python
df[['SMA50', 'SMA200', 'Position']].plot(figsize = (12, 6), secondary_y = 'Position', fontsize = 15)
plt.show()
```

```python
df['Strategy_ret'] = df['Position'].shift() * df['DJI_Return']
df['Strategy'] = df.Strategy_ret.add(1, fill_value = 0).cumprod() * df.iloc[0, 0]
```

```python
df[['DJI_Close', 'Strategy']].plot(figsize = (12, 6), fontsize = 15)
plt.title('SMA50 vs SMA200', fontsize = 20)
plt.legend(fontsize = 15)
plt.show()
```

##Contrarian SMA

```python
df['Position'] = -np.sign(df.SMA50.sub(df.SMA200))
df['Strategy_ret'] = df['Position'].shift() * df['DJI_Return']
df['Strategy'] = df.Strategy_ret.add(1, fill_value = 0).cumprod() * df.iloc[0, 0]

df[['DJI_Close', 'Strategy']].plot(figsize = (12, 6), fontsize = 15)
plt.title('Contrarian SMA50 vs SMA200', fontsize = 20)
plt.legend(fontsize = 15)
plt.show()
```
