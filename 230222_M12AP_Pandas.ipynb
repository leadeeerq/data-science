#Filtering DataFrame with multiple conditions

##AND

```python
mask4 = titanic.sex == 'male'
mask4.head()
```

```python
mask5 = titanic.age > 14
mask5.head()
```

```python
(mask4 & mask5).head()
```

```python
male_survived = titanic.loc[mask4 & mask5, ['survived', 'pclass', 'sex', 'age']]
male_survived.head()
```

```python
male_survived.info(), male_survived.describe()
```

##OR

```python
mask1 = titanic.sex == 'female'
mask2 = titanic.age < 14
(mask1 | mask2).head()
```

```python
woman_or_child = titanic.loc[mask1 | mask2, ['survived', 'pclass', 'sex', 'age']]
woman_or_child.head()
```

```python
woman_or_child.describe()
```

```python
woman_child = titanic.loc[mask1 & mask2, ['survived', 'pclass', 'sex', 'age']]
woman_child.head()
```

```python
woman_child.describe()
```

##User-defined function

```python
def range(series):
  return series.max() - series.min()
```

```python
sales.apply(range, axis = 0)
```

##apply()

```python
sales.apply(range, axis = 1)
```

```python
sales.apply(lambda series: series.max() - series.min(), axis = 1)
```
