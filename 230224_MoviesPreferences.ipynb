##The best and the worst movies

```python
from IPython.display import HTML
```

```python
df_best = df[['poster_path', 'title', 'budget_musd', 'revenue_musd', 'vote_count', 'vote_average', 'popularity']].copy()
df_best
```

```python
df_best['Profit'] = df['revenue_musd'].sub(df['budget_musd'])
df_best['ROI'] = df['revenue_musd'].div(df['budget_musd'])
```

```python
df_best.columns = ['', 'Title', 'Budget', 'Revenue', 'Votes', 'Average Rating', 'Popularity', 'Profit', 'ROI']
```

```python
df_best.set_index('Title', inplace = True)
df_best
```

###fillna()

```python
df_best['Budget'].fillna(0, inplace = True)
df_best['Votes'].fillna(0, inplace = True)
```

```python
df_best.info()
```

###Definition of the function

```python
def best_worst(n, by, ascending = False, min_budget = 0, min_votes = 0):
  df2 = df_best.loc[(df_best['Budget'] >= min_budget) & (df_best['Votes'] >= min_votes), ["", 'Average Rating', by]].sort_values(by = by, ascending = ascending).head(n).copy()
  return HTML(df2.to_html(escape = False))
```

```python
best_worst(5, 'ROI', True, 5, 100)
```

###The best and the worst movies

```python
best_worst(n = 5, by = 'Revenue')
```

```python
best_worst(5, 'Budget')
```

```python
best_worst(5, 'Profit')
```

```python
best_worst(5, 'Profit', ascending = True)
```

```python
best_worst(5, 'Average Rating', min_votes = 20, min_budget = 1)
```

###stack() to generate rows for actors

```python
act.stack().reset_index(level = 1, drop = True).to_frame()
act = act.stack().reset_index(level = 1, drop = True).to_frame()
```

```python
act
```

```python
act.columns = ['Actor']
act
```

###Merging dfs (act df in the left side)

```python
act = act.merge(df[['title', 'revenue_musd', 'vote_average', 'popularity']], how = 'left', left_index = True, right_index = True)
```

```python
act
```

###Label aggregation - agg()

```python
agg = act.groupby('Actor').agg(Total_Revenue = ('revenue_musd', 'sum'),
                               Mean_Revenue = ('revenue_musd', 'mean'),
                               Mean_Rating = ('vote_average', 'mean'),
                               Mean_Popularity = ('popularity', 'mean'),
                               Total_Movies = ('Actor', 'count'))
```
