##Importing and saving the movies dataset

```python
import json
```

```python
movie_id = [0, 260513, 321612, 299534, 19995, 109445, 351286, 330457, 140607, 299536, 181808, 12445, 597, 135397]
```

```python
basic_url = 'https://api.themoviedb.org/3/movie/{}?{}'
```

```python
json_list = []
for id in movie_id:
  url = basic_url.format(id, api_key)
  r = requests.get(url)
  if r.status_code != 200:
    continue
  else:
    data = r.json()
    json_list.append(data)
df = pd.DataFrame(json_list)
```

```python
requests.get(basic_url.format(0, api_key)).status_code
```

```python
df
```

###Selecting a subset of columns

```python
df = df.loc[:, ['title', 'id', 'revenue', 'genres', 'belongs_to_collection', 'runtime']].sort_values(by = 'revenue', ascending = False)
```
